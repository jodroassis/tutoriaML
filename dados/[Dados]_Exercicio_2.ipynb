# An√°lise Estat√≠stica
Qual foi o total, o m√°ximo, o m√≠nimo do objetivo financeiro de todos os projetos (goal)? <br>
Qual o total, o m√°ximo, o m√≠nimo e o valor que representa 25% do dinheiro levantado (pledged)? <br>
Qual o total, o m√°ximo, o m√≠nimo e o valor que representa 75% dos apoiadores dos projetos(backers)? <br>

```python
# 1.
total_goal = df['goal'].sum()
maximo_goal = df['goal'].max()
minimo_goal = df['goal'].min()
print(f'Total: {total_goal}, M√°ximo: {maximo_goal}, M√≠nimo: {minimo_goal}')

# 2.
total_pledged = df['pledged'].sum()
maximo_pledged = df['pledged'].max()
minimo_pledged = df['pledged'].min()
vinte_cinco = df['pledged'].quantile(0.25)
print(f'Total: {total_pledged}, M√°ximo: {maximo_pledged}, M√≠nimo: {minimo_pledged}, 25%: {vinte_cinco}')

# 3.
total_backers = df['backers'].sum()
maximo_backers = df['backers'].max()
minimo_backers = df['backers'].min()
setenta_cinco = df['backers'].quantile(0.75)
print(f'Total: {total_backers}, M√°ximo: {maximo_backers}, M√≠nimo: {minimo_backers}, 75%: {setenta_cinco}')
```
