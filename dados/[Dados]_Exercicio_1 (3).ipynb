##8. Transforme as variaveis categóricas de linguagem e genero em númericas (utilize dummy)

```python
dummies_linguagem = pd.get_dummies(dados['Original_Language'], prefix='Lang', drop_first=True)

dummies_genero = dados['Genre'].str.get_dummies(sep=',')

# 'axis=1' significa que estamos concatenando colunas
dados_numericos = pd.concat([dados, dummies_linguagem, dummies_genero], axis=1)

# Remove as colunas de texto originais que não são mais necessárias
dados_numericos = dados_numericos.drop(['Original_Language', 'Genre'], axis=1)

# Mostra as primeiras linhas do resultado
print("DataFrame com variáveis Dummies:")
display(dados_numericos.head())
```

##9. Normalize as variaveis numéricas

```python
from sklearn.preprocessing import StandardScaler
colunas_para_escalar = ['Vote_Count', 'Vote_Average']

# Copia
df_padronizado = dados_numericos.copy()

# 2. Inicializa o scaler
scaler_std = StandardScaler()

# 3. Padroniza numéricas
df_padronizado[colunas_para_escalar] = scaler_std.fit_transform(df_padronizado[colunas_para_escalar])

# 4. Verifique o resultado
print("\n--- DataFrame Padronizado (Standard) ---")
display(df_padronizado[colunas_para_escalar].describe())
# display(df_padronizado.head())

#Salvar
dados_processados = df_padronizado.copy()
dados_processados.to_csv('dados_processados.csv', index=False)
print("Dataset processado salvo localmente.")
```

##10. Armazene esses valores como um artefato dentro do **MLFlow**

```python
# Salvar o dataset processado como um arquivo CSV
dados_processados = "dados_processados.csv"
dados.to_csv(dados_processados, index=False)
print("Dataset processado salvo localmente.")

# Registrar o dataset processado como artefato no MLflow
mlflow.start_run()  # Iniciar um novo experimento
mlflow.log_artifact(dados_processados)  # Registrar o arquivo como artefato
mlflow.end_run()  # Encerrar o experimento

print("Features armazenadas e versionadas com sucesso no MLflow!")
```

##11. QUais insights é possivel obter desses dados?

```python
#Quais Generos são mais populares?
media_popularidade = dados.groupby('Genre')['Popularity'].mean()
print("Média de Popularidade por Gênero:")
print(media_popularidade.sort_values(ascending=False).head(10))

#Qual o filme mais popular?
filme_mais_popular = dados.loc[dados['Popularity'].idxmax()]
print("\nFilme Mais Popular:")
print(filme_mais_popular)
```
